```python
import cv2
from matplotlib import pyplot as plt
```

```python
# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h #tomonlar nisbati. 0 va 1 orasiga chiqadi. 
    plt.figure(figsize=(size * aspect_ratio,size)) #aspekt ration saqlab qolgan holatda display qilish 
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

# Download and unzip our images. If Google  colab 
# !wget https://moderncomputervision.s3.eu-west-2.amazonaws.com/images.zip
# !unzip -qq images.zip
```

```python
#input imageni yuklash 
image = cv2.imread('Datasets/images/castara.jpeg')
imshow('Castra oroli', image)
```

```python
# cvtColor funksiyasidan foydalanamiz 
# 2 ta argument oladi,  birinchisi input rasm
# ikkinchisi qaysi format color spacega o'tishligi
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow("Converted to Grayscale", gray_image)
```
