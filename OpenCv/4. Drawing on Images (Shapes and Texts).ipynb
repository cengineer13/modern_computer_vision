## **Drawing images and shapes using OpenCV**

Firsty, let's import OpenCV and numpy and define our imshow function

````python
# Our Setup, Import Libaries, Create our Imshow Function and Download our Images
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
````

### Make a blank square

````python
# Create a black image using numpy to create and array of black 
image = np.zeros((512,512,3), np.uint8)

# Can we make this in black and white? grayscale
image_gray = np.zeros((512,512), np.uint8)

# Black would be the same as a greyscale or color image (same for white)
imshow("Black Canvas - RGB Color", image, size=5)
imshow("Black Canvas - Grayscale", image_gray,size=5)
````

### **DRAW LINE**

```cv2.line(image, starting cordinates, ending cordinates, color, thickness)```

````python
#cv2.line funksiya inplace funksiya hisoblanadi. Ya'ni rasmni ozgartirib qaytaradi
#Black canvasga line chizish
cv2.line(image, (0,0), (511,511), (255,255,255), 5)
imshow('Black canvas with Diagonal line', image)
````

### **DRAW RECTANGLE**

```cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)```

````python
# Blank canvas yana bir bor  chizamiz. Chunki eskisiga allaqachon line chizilgan 
image = np.zeros((512,512,3), np.uint8)

# Agar thicknes argument positiv bolsa qalingili rectangle tomonini. 
#Negative bo'lsa rectangle ichi to'ldiriladi
cv2.rectangle(image, (120,100), (250,250), (127,50,127), 10)
imshow("Black Canvas With Pink Rectangle", image)
````

### **DRAW CIRCLE**

```cv2.cirlce(image, center, radius, color, fill)```

````python
image = np.zeros((512,512,3), np.uint8)

cv2.circle(image, (350, 350), 100, (255,255,255), -1)  #thicknes -1 fill circle inside
imshow("Black Canvas With Green Circle", image)
````

### **DRAW POLYGONS - KO"PBURCHAKLI FIGURALAR**

```cv2.polylines(image, points, Closed?, color, thickness)```

if Closed = True, we join the first and last points.

````python
image = np.zeros((512,512,3), np.uint8)

pts = np.array([[10,40],[100,200], [255,19], [300,250]]) #polygon uchun kordinatalar. matrix tuzib olyapmiz
pts = pts.reshape(-1,1,2)  #reshape orqali 3dimensionga otkazib olyapmiz 
cv2.polylines(image, [pts], True, (0,0,255), 10) 
imshow("Polygon on image",image, size=5)
````

**Note** cv2.polylines requires our data be shaped in this way:

### **ADDING TEXT with cv2.putText**

```cv2.putText(image, 'Text to Display', bottom left starting point, Font, Font Size, Color, Thickness)```

**Available Fonts**
- FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN
- FONT_HERSHEY_DUPLEX,FONT_HERSHEY_COMPLEX 
- FONT_HERSHEY_TRIPLEX, FONT_HERSHEY_COMPLEX_SMALL
- FONT_HERSHEY_SCRIPT_SIMPLEX
- FONT_HERSHEY_SCRIPT_COMPLEX`

````python
image = np.zeros((1000,1000,3), np.uint8)

text = "Dusmatov M"
cv2.putText(image, text, (200,500), cv2.FONT_HERSHEY_COMPLEX,3, (0,255,255))
imshow("Text written image", image)
````
