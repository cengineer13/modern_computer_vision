### **Loading Images**

```python
#imread funksiyasi orqali pathni bergan holda rasmni yuklash
image = cv2.imread('Datasets/images/castara.jpeg')
```

### **Displaying Images**

```python
from matplotlib import pyplot as plt

#matplotlib kutubxonasi orqali ekranga chiqarish 
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#Kichikroq funksiya matplotlibda ekranga chiqarish uchun
def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
imshow('Bu orol rasmi', image)
```

### **Saving Images**

```python
# imwrite funksiyasi va path+format kiritilishi yordamida rasmni saqlaydi
#jpg format
cv2.imwrite('output.jpg', image)
```

```python
#PNG ((Portable Network Graphics) yo'qotilmaydigan bitmap formatda saqlash
cv2.imwrite('output.png', image)
```
