# **Openning Videos in OpenCV**

**In this lesson we'll learn:**
1. Open a video using cv2.VideoCapture()

**NOTE:**

AVI video formats are universally supported by all versions of OpenCV so if you're prioritizing cross platform compatibility use .AVI files.

**Other commonly suppported Video Codecs**
1. MPEG4 or MP4
2. FLV

The full list of codecs you may use on a system depends on just what one you have installed.

````python
import cv2

cap = cv2.VideoCapture('Datasets//videos/drummer.mp4')

while cap.isOpened():
    
    ret, frame = cap.read()
    
    # if frame is read correctly ret is True
    if not ret:
        print("Stream stopped.")
        break

    cv2.imshow('Video', frame)
    
    if cv2.waitKey(1) == ord('q'):
        break
    
        
cap.release()
cv2.destroyAllWindows()
````

### Check Video Streams before processing

````python
import cv2

cap = cv2.VideoCapture('Datasets//videos/drummer.mp4') 

if not cap.isOpened(): 
    print("Cannot open camera or video file") 
    exit() #exit program. in jupyter notebook stop kernel

while cap.isOpened(): 
    
    ret, frame = cap.read()
    
    #ret is true when video stream is reads correctly
    if not ret: 
        print("Stream stopped..")
        break
    
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(1) == ord('q'):
        break
    
        
cap.release()
cv2.destroyAllWindows()
````
