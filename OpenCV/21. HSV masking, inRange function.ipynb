#### **Remember The HSV Color Space**

![](https://answers.opencv.org/upfiles/15186766673210035.png)

- Hue: 0 - 179 
- Saturation: 0 - 255
- Value (Intensity): 0 - 255

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('Datasets/images/truck.jpg')

#define blue color in HSV
lower = np.array([90,0,0])
upper = np.array([135,255,255])

#convert from bgr to hsv in order to easily filter
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

#Use inRange function to capture only values between lower and upper
mask = cv2.inRange(hsv_image, lower, upper) 

# Perform Bitwise AND on mask and our original frame
filtered = cv2.bitwise_and(image, image, mask=mask)

imshow('truck', image)
imshow('hsv image', hsv)
imshow('mask', mask) #white are is masked
imshow('Filtered', filtered)
```
