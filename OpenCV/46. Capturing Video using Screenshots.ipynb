# 2. **Capture Video from Screen**

```python
import numpy as np
# copy the contents of the screen or the clipboard to a PIL image memory.
from PIL import ImageGrab #screenshot module 
import cv2
import time

last_time = time.time()

while(True):
    
    #PIL imageni np arrayga otkazib olyapmiz. 
    frame = np.array(ImageGrab.grab(bbox=(0,0,500,500))) #bbox olchami
    
    # Get fullscreen
    #frame = np.array(ImageGrab.grab())
    
    # Display FPS rate
    FPS = 1.0 / (time.time()-last_time)
    print('FPS = {}'.format(FPS))
    last_time = time.time()
    
    # Display Screen
    cv2.imshow('window', cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    
    if cv2.waitKey(1) == ord('q'): #q is the Enter Key
        print("Exited...")
        break
        
cv2.destroyAllWindows()
```

# 3. Use MSS library - super fast frame rate
https://python-mss.readthedocs.io/index.html

```python
import numpy as np
import cv2
from mss import mss
from PIL import Image
import time
```

```python
#monitor properties to grab screen
bounding_box = {'top': 0, 'left': 0, 'width': 400, 'height': 400}

sct = mss()
frame_count = 0 
last_time = time.time()

while True: 
    frame_count += 1
    screenshot = sct.grab(bounding_box) 
    cv2.imshow('screen', np.array(screenshot)) 
    
    # Display FPS rate
    if frame_count % 30 == 0:
        FPS = 1.0 / (time.time()-last_time)
        print('FPS = {}'.format(FPS))
    last_time = time.time()    
    
    if cv2.waitKey(1) == ord('q'): #q is the Enter Key
        print("Exited...")
        break
        
cv2.destroyAllWindows()
```

### Video screen record with PyAutoGui 

```python
SCREEN_SIZE = (1920, 1080)
#time.time hozirgi vaqtni ifodalaydi yani 3 minut keyin digani 60*3 = 60 sec * 3 = 3 minuttes
TIME_END = time.time() + 60 * 3

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('Datasets/videos/output/screen_recorded.avi', fourcc, 20.0, (SCREEN_SIZE))

fps = 120
prev = 0 #slowing down video process
i = 0 
while time.time() <TIME_END: 
    
    time_elapsed = time.time() - prev
    
    screenshot = pyautogui.screenshot()
    if time_elapsed > 1.0/fps: 
        prev = time.time()
        frame = cv2.cvtColor(np.array(screenshot), cv2.COLOR_BGR2RGB) 
        
        print("Screen recording is started.....") if i == 0 else False
        
    
        out.write(frame)
    i+=1    
    cv2.waitKey(100)
out.release()
print("Screen recording is stopped....")
cv2.destroyAllWindows()
```
