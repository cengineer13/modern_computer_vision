# **Color Spaces**

In this lesson we'll learn to:
1. View the individual channels of an RGB Image
2. Manipulate a color space
3. Introduce HSV Color Spaces

```python
# Our Setup, Import Libaries, Create our Imshow Function and Download our Images
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

# Download and unzip our images - Colab uchun
# !wget https://moderncomputervision.s3.eu-west-2.amazonaws.com/images.zip
# !unzip -qq images.zip
```

```python
# Rasmnni yuklash
image = cv2.imread('Datasets/images/castara.jpeg')

#split funksiyasi channelarni alohida tuple qilib rasm
B, G, R = cv2.split(image)
print(B.shape)
print(G.shape)
print(R.shape)
```

```python
#Har bir color space o'zi yakka huddi grayscale kabi bo'ladi. 0 dan 255 gacha intensity faqat o'sha rangi
imshow("Blue Channel Only", B)
```

```python
import numpy as np

#1 chanelli yani h x w matrissali rasm yaratib oldik qora rasm
zeros = np.zeros(image.shape[:2], dtype = "uint8") 

#
imshow("Red", cv2.merge([zeros, zeros, R]))
imshow("Green", cv2.merge([zeros, G, zeros]))
imshow("Blue", cv2.merge([B, zeros, zeros]))
```

```python
image = cv2.imread('Datasets/images/castara.jpeg')

B, G, R = cv2.split(image)

# Rasmni cv2.merge funksiyasi yordamida 1ta channeli rasmlarni tiklab ko'ramiz
merged = cv2.merge([B, G, R]) 
imshow("Merged", merged) 
```

```python
#Keling yashil rangni kuchaytirib ko'ramiz. 
merged = cv2.merge([B, G+100, R]) #yashil rang matrixdagi har bir qiymatga 100 ni qoshib rangni intensivligini oshiradi
imshow("Kuchaytirilgan ko'k rang", merged)
```

## **The HSV Color Space**

- Hue: 0 - 179 
- Saturation: 0 - 255
- Value (Intensity): 0 - 255


![](https://upload.wikimedia.org/wikipedia/commons/f/f2/HSV_color_solid_cone.png)

```python
# Reload our image
image = cv2.imread('Datasets/images/castara.jpeg')

# Convert to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_image)
```

### **Let's view each channel type in the HSV Color Space representation**

```python
imshow('Hue', hsv_image[:,:,0])
imshow('Saturation', hsv_image[:,:,1])
imshow('Value', hsv_image[:,:,2])
```
